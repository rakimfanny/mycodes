## we need to tag the customers still active in the last deciscive month , in order to determine further , who churned or not

```python
view_customer=churn[['total_og_mou_9','total_ic_mou_9','vol_2g_mb_9','vol_3g_mb_9']]
```

```python
view_customer.head()
```

### we put the number '0 where there is not activities observed , and '1' otherwise

```python
for c in view_customer:
    view_customer[c]=np.where(view_customer[c]!=0,1,view_customer[c])
    view_customer[c]=np.where(view_customer[c]==0,0,view_customer[c])
```

```python
view_customer.head()
```

```python
view_customer=view_customer.astype(int)
```

```python
view_customer.head()
```

### now we add row-wise the digit to know who churned or not , '0' will designate who churned ans '1' who did not 

```python
churned=view_customer.apply(lambda x : x['total_og_mou_9']+x['total_ic_mou_9']+x['vol_2g_mb_9']+x['vol_3g_mb_9'],axis=1)
```

```python
for c in churned:
    churned[c]=np.where(churned[c]!=0,1,churned[c])
    churned[c]=np.where(churned[c]==0,0,churned[c])
```
